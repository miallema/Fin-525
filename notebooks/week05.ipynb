```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
spy1 = pd.read_csv('../data/SPY.P_events_20161011.csv.gz', compression = 'gzip')
```

```python
spy1.head()
```

```python
def excel_to_date(ordinal):
    excel = datetime.datetime(1900,1,1,0,0).toordinal()
    excel_bug = 2
    return pd.Timestamp.fromordinal(excel-excel_bug+ordinal)
```

```python
spy1.index = spy1['date'].apply(excel_to_date) + pd.to_timedelta(spy1['second'],unit = 's')
spy1 = spy1.drop(columns = ['date', 'second'])
spy1.index.name = 'date'
spy1 = spy1.between_time('09:30:00','16:00:00')
spy1 = spy1.tz_localize('America/New_York',level = 0)
```

```python
apple = pd.read_csv('../data/AAPL.OQ_events_20161011.csv.gz', compression = 'gzip')
ubs = pd.read_csv('../data/UBSG.S_events_20161011.csv.gz', compression = 'gzip')
```

```python
apple.index = apple['date'].apply(excel_to_date) + pd.to_timedelta(apple['second'],unit = 's')
apple = apple.drop(columns = ['date', 'second'])
apple.index.name = 'date'
apple = apple.between_time('09:30:00','16:00:00')
apple = apple.tz_localize('America/New_York',level = 0)
apple.columns = ['bid_apple','bidQ_apple','ask_apple','askQ_apple','last_apple','lastQ_apple']
```

```python
apple.head()
```

```python
ubs.index = ubs['#date_xl'].apply(int).apply(excel_to_date) + pd.to_timedelta(ubs[' ts'],unit = 's')
ubs = ubs.drop(columns = ['#date_xl', ' ts'])
ubs.index.name = 'date'
ubs = ubs.between_time('09:30:00','16:00:00')
ubs = ubs.tz_localize('Europe/Zurich',level = 0)
ubs.columns = ['bid_ubs','bidQ_ubs','ask_ubs','askQ_ubs','last_ubs','lastQ_ubs']
```

```python
ubs.head()
```

```python
d = apple.join(ubs, how = 'outer')
#d.iloc[0,0]= 117.65
#d = d.fillna(method = 'pad')
d['bid_apple_normalized'] = d['bid_apple']/apple['bid_apple'][0]
d['bid_ubs_normalized'] = d['bid_ubs']/ubs['bid_ubs'][0]
```
